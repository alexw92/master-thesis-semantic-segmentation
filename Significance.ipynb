```python
import pandas as pd
from scipy.stats import wilcoxon
import numpy as np
```

```python
first_col = ['score']
cols = ['other', 'building', 'wood', 'water', 'road']
```

```python
def parse(x):
    try:
        return float(x)
    except:
        return float(0.0)
    
def parse_df(df):
    l = df.shape[0]
    for col in cols:
        df[col] = df[col].apply(parse).fillna(0.0)
    df.dropna(inplace=True)
    print("Rows reduced from {} to {}".format(l, df.shape[0]))
```

```python
de = pd.read_csv("C:/Uni/Masterstudium/ma-werthmann/code/analysis/segnet/detop15_nores", header=None)
de.columns = first_col + cols
parse_df(de)
de.head()
```

```python
eu = pd.read_csv("C:/Uni/Masterstudium/ma-werthmann/code/analysis/segnet/eutop25_nores", header=None)
eu.columns = first_col + cols
parse_df(eu)
eu.head()
```

```python
def test_for_column(column, score='iou'):
    d = de[de.score==score][column].values
    e = eu[eu.score==score][column].values

    l = min(len(e), len(d))

    return wilcoxon(d[:l], e[:l])
```

```python
for score in ['iou', 'acc']:
    for column in cols:
        print("p-value for score {} and column {} is {}".format(score, column, test_for_column(column, score)[1]))
```
